## zadanie 1
Gęstość powietrza $ρ$ zmienia się z wysokością $h$ w następujący sposób:

$$
\def\arraystretch{1.5}
   \begin{array}{l|l|l|l}
   \hline \hline
   h (\text{km})& 0 & 3 & 6 \\ \hline
\rho (\text{kg/m}^3) &  1,225 & 0,905 & 0,652 \\ \hline \hline
\end{array}
$$
Wyraź $ρ(h)$ jako funkcję kwadratową w $h$.

```python
points = [[0, 1.225], [3, 0.905], [6, 0.652]]
def poly(x, coeffs):
    result = 0
    ln = len(coeffs)
    for i, coeff in enumerate(coeffs):
        pow = ln - 1 - i
        result += coeff * x**pow
    return result
def interpolation(points, example = 1):
    x = np.array([p[0] for p in points])
    y = np.array([p[1] for p in points])
    n = len(x)-1
    W = np.copy(y)
    # obliczanie ilorazów różnicowych
    for i in range(1,n+1):
        for j in range(n,i-1,-1):
            W[j]=(W[j]-W[j-1])/(x[j]-x[j-i])
    a = np.copy(W)
    # zmiana ilorazów różnicowych na współczynniki wielomianu
    for i in range(n):
        for j in range(n-1,i-1,-1):
            a[j]=a[j]-x[j-i]*a[j+1]
    if example == 1:
        xs = np.linspace(min(x)-1, max(x)+1, 100)
        # obliczanie wartości wielomianu na wybranym przedziale
        ys = [poly(s, a[::-1]) for s in xs]
        # przedstawienie na wykresie
        plt.scatter(x,y, color = 'teal', label = "dane")
        plt.plot(xs, ys, color = 'lightcoral', label = 'interpolacja')
        plt.xlabel('h')
        plt.ylabel(r'$\rho$')
        plt.legend()
        return a
    if example == 3:
        return [poly(s, a[::-1]) for s in [5, 50 , 5000]]
interpolation(points)
```

## zadanie 4
Wyznacz parametry a i b tak, aby funkcja $f(x) = ae^{bx}$ aproksymowała
poniższe dane w sensie najmniejszych kwadratów.
$$
\def\arraystretch{1}
   \begin{array}{l|llllll}
   \hline \hline
   x & 1,2 & 2,8 & 4,3 & 5,4 & 6,8 & 7,9\\ \hline
   y & 7,5 & 16,1 & 38,9 & 67,0 & 146,6 & 266,2 \\ \hline \hline
\end{array}
$$
Policz odchylenie standardowe. Wyniki przedstaw na wykresie.

```python
x = np.array([1.2, 2.8, 4.3, 5.4, 6.8, 7.9])
y = np.array([7.5, 16.1, 38.9, 67.0, 146.6, 266.2])
```

**metody biblioteczne**

```python
from scipy.optimize import curve_fit

def model(x, a, b):
    return a*np.exp(b*x)

popt, _ = curve_fit(model, x, y)
a1, b1 = popt
fun1 = lambda x: a1 * np.exp(b1 * x)
residuals1 = y - fun1(x)
std1 = np.std(residuals1)
X_fit = np.linspace(min(x), max(x), 100)
Y_fit = model(X_fit, *popt)
plt.scatter(x, y, label = 'dane', color='teal')
plt.plot(X_fit, Y_fit, label=f'y = {a1:.3f} * exp({b1:.3f} * x)', color='lightcoral')
plt.legend()
print(f'a = {a1}, b = {b1}')
print(f'y = {a1:.3f} * exp({b1:.3f} * x)')
print(f'Odchylenie standardowe = {std1}')
```

```python
def exp_approx(x, y):
    # nakładając logarytm uzyskamy
    # log(y) = log(a) + bx
    coeff = np.polyfit(x, np.log(y), 1)
    a, b = np.exp(coeff[1]), coeff[0]
    return a, b

a2, b2 = exp_approx(x, y)
fun2 = lambda x: a2 * np.exp(b2 * x)
residuals2 = y - fun2(x)
std2 = np.std(residuals2)
print(f'a = {a2}, b = {b2}')
print(f'y = {a2:.3f} * exp({b2:.3f} * x)')
print(f'Odchylenie standardowe = {std2}')
Y_fit2 = fun2(X_fit)
plt.scatter(x, y, label = 'dane', color='teal')
plt.plot(X_fit, Y_fit2, label=f'y = {a2:.3f} * exp({b2:.4f} * x)', color='lightcoral')
plt.legend()
```

Zadanie zostało wykonane na 2 sposoby:

1. Używając `curve_fit` i modelu eksponencjalnego. Uzyskane wyniki to: \
a = 3.613733933422907, b = 0.5442486608957392 \
y = 3.614 * exp(0.544 * x) \

Wyniki są mocno zbliżone do siebie, jednka możemy zauważyć, że odchylenie standardowe modelu od danych jest mniejsze w 1. przypadku. Możemy więc wnioskować, że dopasowanie funkcji wykładnicze daje lepsze rezultaty od dopasowywania jej przekształceń (nałożenie logarytmu).

**implemetacje własne** \
1) regresja logarytmiczna

```python
import numpy as np
import matplotlib.pyplot as plt

# Dane
# x = np.array([1.2, 2.8, 4.3, 5.4, 6.8, 7.9])
# y = np.array([7.5, 16.1, 38.9, 67.0, 146.6, 266.2])
def approx_2(x, y):
    ln_y = np.log(y)
    x_mean = np.mean(x)
    ln_y_mean = np.mean(ln_y)
    b = np.sum((x - x_mean) * (ln_y - ln_y_mean)) / np.sum((x - x_mean)**2)
    ln_a = ln_y_mean - b * x_mean
    a = np.exp(ln_a)
    return a, b
a3, b3 = approx_2(x, y)

# Wykres dopasowania
Y_fit3 = a3 * np.exp(b3 * X_fit)

plt.scatter(x, y, label='Dane', color='teal')
plt.plot(X_fit, Y_fit3, label=f'y = {a3:.3f} * exp({b3:.3f} * x)', color='lightcoral')
plt.legend()
plt.show()
fun3 = lambda x: a3 * np.exp(b3 * x)
residuals3 = y - fun3(x)
std3 = np.std(residuals3)
print(f'a = {a3}, b = {b3}')
print(f'y = {a3:.3f} * exp({b3:.3f} * x)')
print(f'Odchylenie standardowe = {std3}')
```

2) najmniejsze kwadraty z pochodnymi

```python
from scipy.optimize import fsolve
# Funkcja zwracająca pochodne cząstkowe
def derivatives(params):
    a, b = params
    e_bx = np.exp(b * x)
    residuals = y - a * e_bx

    # Pochodne cząstkowe
    dE_da = -2 * np.sum(residuals * e_bx)
    dE_db = -2 * np.sum(residuals * a * x * e_bx)

    return [dE_da, dE_db]

# Rozwiązanie układu równań
initial_guess = [1.0, 1.0]  # Początkowe wartości a i b
solution = fsolve(derivatives, initial_guess) # rozwiązuje układ równań
a4, b4 = solution
Y_fit4 = a4 * np.exp(b4 * X_fit)

plt.scatter(x, y, label='Dane', color='teal')
plt.plot(X_fit, Y_fit4, label=f'y = {a4:.3f} * exp({b4:.3f} * x)', color='lightcoral')
plt.legend()
plt.show()
fun4 = lambda x: a4 * np.exp(b4 * x)
residuals4 = y - fun4(x)
std4 = np.std(residuals2)
print(f'a = {a4}, b = {b4}')
print(f'y = {a4:.3f} * exp({b4:.3f} * x)')
print(f'Odchylenie standardowe = {std4}')
```

## zadanie 5
Kinematyczna lepkość wody $µ_k$ zmienia się z temperaturą $T$ według poniższej tabeli. Wyznacz wielomian trzeciego stopnia, który najlepiej aproksymuje te dane i skorzystaj z niego do obliczenia $µ_k$ dla $T = 10\degree ,\ 30\degree,\ 60\degree \text{ i} \ 90\degree$.
$$
\def\arraystretch{1}
   \begin{array}{l|lllllll}
   \hline \hline
   T (\degree C) & 0 & 21,1 & 37,8 & 54,4 & 71,1 & 87,8 & 100\\ \hline
   \mu_k (10^{-3}m^2/s) & 1,79 & 1,13 & 0,696 & 0,519 & 0,338 & 0,321 & 0,296 \\ \hline \hline
\end{array}
$$

**Rozwiązanie:** \
Szukamy wielomianu postaci $y = ax^3 + bx^2 + cx + d$. \
Chcemy znaleźć takie $a, \ b, \ c, \ d$, które minimalizują błąd:
$$\mathbb{E}(a,b,c,d) = \sum_{i=1}^n (y_i - (ax_i^3 + bx_i^2 + cx_I + d))^2 $$
W postaci macierzowej mamy:
$$\mathit{A} \cdot \beta  = \mathit{Y},$$
$\mathit{A}$ to macierz układów [$x_i^3, x_i^2, x_i, 1$] \
$\beta$ = [$a,b,c,d$] \
$\mathit{Y}$ to wektor wartośći $y$ \
Współczynnik $\beta$ znajdziemy rozwiązując układ:
$$A^TA \cdot \beta = A^TY$$

```python
T = np.array([0, 21.1, 37.8, 54.4, 71.1, 87.8, 100])
mu_k = np.array([1.79, 1.13, 0.696, 0.519, 0.338, 0.321, 0.296])
```

```python
A = np.vstack([T**3, T**2, T, np.ones_like(T)]).T
AtA = np.dot(A.T, A)
AtY = np.dot(A.T, mu_k)
beta = np.linalg.solve(AtA, AtY)
a, b, c, d = beta
print(f"Współczynniki: \na = {a}, \nb = {b}, \nc = {c}, \nd = {d}")

ts = np.linspace(min(T), max(T), 100)
mu_ks = a * ts**3 + b * ts**2 + c * ts + d

plt.scatter(T, mu_k, color='teal', label='Dane')
plt.plot(ts, mu_ks, color='lightcoral', label=f'{a:.5f}x^3 + {b:.5f}x^2 + {c:.5f}x + {d:.5f}')
plt.xlabel('T')
plt.ylabel('mu_k')
plt.legend()
plt.show()
```

## zadanie 6
 Dopasuj funkcję liniową i kwadratową do danych.
 $$
\def\arraystretch{1}
   \begin{array}{l|llllllll}
   \hline \hline
   x & 1,0 & 2,5 & 3,5 & 4.0 & 1,1 & 1,8 & 2,2 & 3,7\\ \hline
   y & 6,008 & 15,722 & 27,13 & 33,772 & 5,257 & 9,549 & 11,098 & 28,828 \\ \hline \hline
\end{array}
$$

```python
x = np.array([1, 2.5, 3.5, 4.0, 1.1, 1.8, 2.2, 3.7])
y = np.array([6.008,15.722,27.13,33.772,5.257,9.549,11.098,28.828])
def regresja_lin(x,y):
    a = np.sum((x - np.mean(x)) * (y))/np.sum((x-np.mean(x))**2)
    b = np.mean(y) - a * np.mean(x)
    return a, b

ar,br = regresja_lin(x,y)
```

```python
plt.plot(x, ar*x+br, color = 'lightcoral', label = 'regresja liniowa')
plt.scatter(x,y, color = 'teal', label = 'dane'); plt.legend()
```

```python
A = np.vstack([x**2, x, np.ones_like(x)]).T

AtA = [[np.sum(A[:, i] * A[:, j]) for j in range(3)] for i in range(3)]
AtY = [np.sum(A[:, i] * y) for i in range(3)]

def gauss_elimination(A, b):
    n = len(b)
    for i in range(n):
        factor = A[i][i]
        for j in range(i, n):
            A[i][j] /= factor
        b[i] /= factor

        for k in range(i + 1, n):
            factor = A[k][i]
            for j in range(i, n):
                A[k][j] -= factor * A[i][j]
            b[k] -= factor * b[i]

    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = b[i] - sum(A[i][j] * x[j] for j in range(i + 1, n))
    return x

beta = gauss_elimination(AtA, AtY)

a6, b6, c6 = beta
print(f"Współczynniki: a={a6}, b={b6}, c={c6}")

x_fit = np.linspace(min(x), max(x), 100)
y_fit = a6 * x_fit**2 + b6 * x_fit + c6

plt.scatter(x, y, color='teal', label='dane')
plt.plot(x_fit, y_fit, color='lightcoral', label=f'{a6:.3f}x^2 + {b6:.3f}x + {c6:.3f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, color='teal', label='Dane')
plt.plot(x_fit, y_fit, color='lightcoral', label=f'{a:.3f}x^2 + {b:.3f}x + {c:.3f}')
plt.plot(x, ar*x+br, color = 'khaki', label = f'{ar:.3f}x + {br:.3f}')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def calculate_sse(y_actual, y_predicted):
    # Convert the input lists to NumPy arrays for vectorized operations
    y_actual = np.array(y_actual)
    y_predicted = np.array(y_predicted)

    # Calculate squared errors and sum them up
    squared_errors = (y_actual - y_predicted) ** 2
    sse = np.sum(squared_errors)

    return sse
```

```python
y1 = ar * x + br
y2 = a6 * x**2 + b6 * x + c6
errors1 = y - y1
errors2 = y - y2
print('odchylenie standardowe: ', np.std(errors1), np.std(errors2))
print('sse: ',calculate_sse(y, y1), calculate_sse(y, y2))
```

Jak widać, aproksymacja funkcją **kwadratową** daje mniejsze odchylenie standardowe błędów oraz mniejszy błąd kwadratowy niż aproksymacja funkcją liniową, a więc możemy powiedzieć, że jest lepsza.
