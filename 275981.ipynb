### zadanie 1

```python
def z1(x):
    return (6-2*x)/(6+4*x+x**2)

def z2(x):
    return (6-4*x + x**2)/(6+x*2)
```

```python
xs = np.linspace(-1, 3, 10**5)
y1 = z1(xs)
y2 = z2(xs)
e = np.exp(-xs)
```

```python
plt.plot(xs, y1, label = 'z1')
plt.plot(xs, y2, label = 'z2', alpha=0.7)
plt.legend()
plt.title('Przybliżenia Padego')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.suptitle('Przybliżenia Padego funkcji exp(-x)')
plt.subplot(1,2,1)
plt.plot(xs, y1, color = 'orange', label = 'z1')
plt.plot(xs, e, color='black', label = 'exp(-x)', linestyle='--')
plt.legend()
plt.subplot(1,2,2)
plt.plot(xs, y2, label = 'z2')
plt.plot(xs, e, color='black', label = 'exp(-x)', linestyle='--')
plt.legend()
plt.show()
```

```python
b1 = abs(y1-e)
b2 = abs(y2-e)
```

```python
plt.figure(figsize=(10,5))
plt.suptitle('Błędy przybliżenia Padego funkcji exp(-x)')
plt.subplot(1,2,1)
plt.plot(xs, b1, color = 'orange', label = 'z1')
plt.legend()
plt.subplot(1,2,2)
plt.plot(xs, b2, label = 'z2')
plt.legend()
plt.show()
```

### zadanie 4

```python
num = '00111111110110011001100110011001'
len(num)

def ieee_to_dec(num):
    sign = int(num[0])
    exponent = num[1:-23]
    mantissa = num[-23:]
    exp_dec = int(exponent, 2)
    man_dec = int(mantissa, 2)/2**23 +1
    return (-1)**sign*np.power(2,(exp_dec-127))*(man_dec)
ieee = ieee_to_dec(num)
```

```python
abs_err = abs(1.7 - ieee)
rel_err = abs(1.7 - ieee)/1.7
print('obliczone przybliżenie:',ieee,'\nbłąd bezwzględny: ',abs_err, '\nbład względny: ', rel_err)
```

### zadanie 5

```python
def poly(x, coeffs = [6, 5, -13, 1, 1]):
    result = 0
    ln = len(coeffs)
    for i, coeff in enumerate(coeffs):
        pow = ln - 1 - i
        result += coeff * x**pow
    return result
```

```python
dx = 0.0001
xs = np.arange(-10, 10, dx)
start = time.time()
ys = poly(xs)
end = time.time()
poly_time = end - start
print(f"czas wykonania programu: {poly_time}")
```

```python
plt.plot(xs, ys)
```
